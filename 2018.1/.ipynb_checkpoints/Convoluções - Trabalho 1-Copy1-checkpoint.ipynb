### Parte 1 - Operaçao de Convoluçao

Uma operação de convolução, dada pela equação abaixo, pode ser realizada a partir da aplicação de uma máscara/filtro convolucional para cada pixel da imagem.  

\begin{align}
g[n] = \sum_{\tau=-\infty}^\infty s[\tau] h[n]-\tau]
\end{align}

Na equação, g é o resultado da operação, s o sinal de entrada e h a máscara convolucional. A Figura abaixo exemplifica a operação para um pixel em uma imagem de dimensão 8x8, utilizando máscara 3x3.

![title](conv.jpg)

Para realizar a operação em Python, implemente a função convolution(X, kernel, stride, padding), em que X é a imagem de entrada, utilizando apenas funçoes básicas da biblioteca Numpy (i.e. função convolve, convolve2d e outras do Numpy não serão permitidas nesse exercício). A função retornará uma matriz out_conv e deverá permitir a utilização ou não padding, assim como stride com valores acima de 1. 

Para o padding, implemente a função zero_pad(X), que preencherá a imagem de entrada com zeros, retornando a matriz X_pad. 



```python
def zero_pad(X):
    X_pad = np.zeros((X.shape[0] + 2, X.shape[1] + 2, X.shape[2]))  
    X_pad[1:-1, 1:-1] = X
    return X_pad
```

```python
def convolution(X,kernel,stride=1, padding=True):
    n_H = int((X.shape[1]-kernel[0].shape[1]+2*padding)/stride+1)
    n_W = int((X.shape[0]-kernel[0].shape[1]+2*padding)/stride+1)
    out = np.zeros((n_W, n_H,len(kernel)))
    #h,w = X.shape
    h = X.shape[1]
    w = X.shape[0]
    if padding:
        X = zero_pad(X)
    else: 
        h = h - 2
        w = w - 2
    for c in range(len(kernel)):
        for i in range(n_W):
            for j in range(n_H):
                #if X[i*stride:i*stride+kernel.shape[0],j*stride:j*stride+kernel.shape[1]].shape == (3,3):
                if ((i*stride+kernel[c].shape[0]) - i*stride == 3) & ((j*stride+kernel[c].shape[1]) - j*stride == 3):
                    r = np.sum(kernel[c]*X[i*stride:i*stride+kernel[c].shape[0],j*stride:j*stride+kernel[c].shape[1],0])
                    g = np.sum(kernel[c]*X[i*stride:i*stride+kernel[c].shape[0],j*stride:j*stride+kernel[c].shape[1],1])
                    b = np.sum(kernel[c]*X[i*stride:i*stride+kernel[c].shape[0],j*stride:j*stride+kernel[c].shape[1],2])
                    out[i,j,c] = r+g+b/255
                if out[i,j,c]>255:
                    out[i,j,c]=255
                elif out[i,j,c] < 0:
                    out[i,j,c] = 0
    return out
```
